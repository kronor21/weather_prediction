```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('D:\weather_forecasting_ML_project\weather.csv')
```

```python
df.head()
```

```python
# X = df[['MinTemp', 'MaxTemp', 'Rainfall' , 'WindSpeed9am' , 'Humidity3pm']]

x_col = list(df.columns)
x_col.remove('RainTomorrow')
x = df[x_col]
y = df['RainTomorrow']
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
lr = LinearRegression()
lr.fit(X_train, y_train)
```

```python
y_pred = lr.predict(X_test)
```

```python
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_pred)
print('MSE:', mse)
print('RMSE:', rmse)
print('R2 Score:', r2)
```

```python
plt.scatter(y_test, y_pred)
plt.xlabel('Actual Values')
plt.ylabel('Predicted Values')
plt.title('Actual vs Predicted Values for Rain Tomorrow')
plt.show()
```
